```python
df['tenure_group'] = pd.cut(
    df['tenure'],
    bins=[-1, 12, 24, 48, 72],
    labels=['0-12', '12-24', '24-48', '48+']
)
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
)

df.columns = df.columns.str.replace('seniorcitizen', 'senior_citizen')
df.columns = df.columns.str.replace('onlinesecurity', 'online_security')
df.columns = df.columns.str.replace('onlinebackup', 'online_backup')
df.columns = df.columns.str.replace('deviceprotection', 'device_protection')
df.columns = df.columns.str.replace('techsupport', 'tech_support')
df.columns = df.columns.str.replace('streamingtv', 'streaming_tv')
df.columns = df.columns.str.replace('streamingmovies', 'streaming_movies')
df.columns = df.columns.str.replace('monthlycharges', 'monthly_charges')
df.columns = df.columns.str.replace('totalcharges', 'total_charges')
df.columns = df.columns.str.replace('internetservice', 'internet_service')
df.columns = df.columns.str.replace('paymentmethod', 'payment_method')
df.columns = df.columns.str.replace('phoneservice', 'phone_service')
df.columns = df.columns.str.replace('multiplelines', 'multiple_lines')
df.columns = df.columns.str.replace('paperlessbilling', 'paperless_billing')
```

```python
df['churn_flag'] = df['churn'].map({'Yes': 1, 'No': 0})

cols = [col for col in df.columns if col not in ['churn', 'churn_flag']]
df = df[cols + ['churn', 'churn_flag']]
```
